```python
plt.figure(figsize=(5,5))
ax = sns.catplot(x="no_older_children",kind='point',data=df,y='Holliday_Package')
```

Number of older children does not seem like a good predictor on deciding whether an employee is likely to take Holiday Package or not as there is approximately equal chance for employess with 4 or less older children to take the package. Employees with number of older children as 5 does not tend to take the package whereas Employees with number of older children as 6 tend to take the package. But there are very few records with number of older children as 5 or 6

```python
plt.figure(figsize=(5,5))
ax = sns.catplot(x="no_young_children",kind='point',data=df,y='Holliday_Package')
```

There is an equal chance for people having no young children to opt or not opt a Holiday Package. But as the number of children increases to 1 and 2, we can see that per cent of people to prefer a Holiday Package is decreasing although there is slight increase in the per cent when it comes to family with 3 young children.

```python
plt.figure(figsize=(5,5))
ax = sns.catplot(x="foreign",kind='count',data=df,hue='Holliday_Package')
```

Most of the foreigners tend to take a Holiday Package wheras most of the non foreigners does not prefer to take a Holiday Package

```python
# Copy all the independent variables into X dataframe
X = df.drop('Holliday_Package', axis=1)

# Copy target variable into the y dataframe. 
y = df[['Holliday_Package']]
```

```python
### Data Split: Split the data into train and test (70:30).
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30 , random_state=1)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
    'max_iter': [20,50,100,150], #5000,2500,3000,2000,1000,1500,500,600
    'solver': ['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga'],#, 'sgd'
    'tol': [0.01,0.1,0.001], 
}

nncl = LogisticRegression(random_state=1,verbose=True)

grid_search = GridSearchCV(estimator = nncl, param_grid = param_grid, cv = 10)
```

```python
grid_search.fit(X_train, y_train)
print(grid_search.best_params_)
best_grid = grid_search.best_estimator_
best_grid
```
