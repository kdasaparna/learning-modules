### Remove all the stopwords from all the three speeches

```python
stopwords = nltk.corpus.stopwords.words('english') +list(string.punctuation)+ ['--']
```

```python
Roosevelt_speech_words = (w.lower() for w in inaugural.words('1941-Roosevelt.txt'))
Cleaned_Roosevelt_speech_words = [word for word in Roosevelt_speech_words if word not in stopwords]
```

```python
Kennedy_speech_words = (w.lower() for w in inaugural.words('1961-Kennedy.txt'))
Cleaned_Kennedy_speech_words = [word for word in Kennedy_speech_words if word not in stopwords]
```

```python
Nixon_speech_words = (w.lower() for w in inaugural.words('1973-Nixon.txt'))
Cleaned_Nixon_speech_words = [word for word in Nixon_speech_words if word not in stopwords]
```

### Which word occurs the most number of times in his inaugural address for each president? Mention the top three words. (after removing the stopwords) 

#### Most common words in Roosevelt speech

```python
nltk.FreqDist(Cleaned_Roosevelt_speech_words).most_common(3)
```

#### Most common words in Kennedy speech

```python
nltk.FreqDist(Cleaned_Kennedy_speech_words).most_common(3)
```

#### Most common words in Nixon speech

```python
nltk.FreqDist(Cleaned_Nixon_speech_words).most_common(3)
```

#### Plot the word cloud of each of the speeches of the variable. (after removing the stopwords) 

```python
pip install wordcloud
```

```python
from wordcloud import WordCloud,STOPWORDS
import matplotlib.pyplot as plt   

words = ' '.join(Cleaned_Roosevelt_speech_words)
cleaned_word = " ".join([word for word in words.split()
                            if 'http' not in word
                                and not word.startswith('@')
                                and word != 'RT'
                            ])
wordcloud = WordCloud(stopwords=STOPWORDS,
                      background_color='black',
                      width=3000,
                      height=2500
                     ).generate(cleaned_word)
plt.figure(1,figsize=(10, 10))
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```
