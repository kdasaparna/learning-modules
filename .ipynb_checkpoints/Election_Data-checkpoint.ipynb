### Encode the data (having string values) for Modelling. Is Scaling necessary here or not? Data Split: Split the data into train and test (70:30).

```python
from sklearn.preprocessing import LabelEncoder   # import label encoder

labelencoder = LabelEncoder()  

df["vote"] = labelencoder.fit_transform(df.vote)
df["gender"] = labelencoder.fit_transform(df.gender)
```

```python
df
```

For the variable vote, the value Labour has been encoded as 1 and Conservative as 0 and for the variable gender male has been encoded as 1 and female has been encoded as 0

### Data Split

```python
X = df.drop("vote", axis=1)

y = df.pop("vote")

X.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, train_labels, test_labels = train_test_split(X, y, test_size=.30, random_state=1)
```

```python
X_train.shape
```

```python
X_test.shape
```
